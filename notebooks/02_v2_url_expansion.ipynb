```python
# Final pool summary
import json
from pathlib import Path
from collections import Counter

candidates_path = Path('../data/raw/wdc_candidate_urls.jsonl')
types = Counter()
sources = Counter()
with open(candidates_path) as f:
    for line in f:
        rec = json.loads(line)
        types[rec['schema_type']] += 1
        sources[rec.get('source', 'wdc')] += 1

print(f'Total candidate URLs: {sum(types.values()):,}')
print(f'\nBy type:')
for t, n in types.most_common():
    print(f'  {t:25s} {n:6,}')
print(f'\nBy source:')
for s, n in sources.most_common():
    print(f'  {s:15s} {n:6,}')
```
