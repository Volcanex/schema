# 01_v2 — WDC URL Extraction

Extracts candidate URLs from Web Data Commons class-specific N-Quads files.
Smart quota-based approach: stops downloading as soon as we have enough URLs per type.

**Output:** `data/raw/wdc_candidate_urls.jsonl`

```python
!python3 -u ../scripts/wdc_extract.py
```

```python
# Quick summary
import json
from pathlib import Path
from collections import Counter

candidates_path = Path('../data/raw/wdc_candidate_urls.jsonl')
types = Counter()
with open(candidates_path) as f:
    for line in f:
        types[json.loads(line)['schema_type']] += 1

print(f'Total URLs: {sum(types.values()):,}\n')
for t, n in types.most_common():
    print(f'  {t:25s} {n:6,}')
```
